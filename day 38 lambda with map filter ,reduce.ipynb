```python
num=[11,23,12,3121,12]
```

```python
#map(function,iterable object)

num=list(map(lambda x: x**2,num))
num
```

```python
num=list(filter(lambda num:num%2==1,num))
num
```

```python
from functools import reduce
reduce(lambda a,b:a+b,num)
```
