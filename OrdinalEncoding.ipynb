```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
data=[
    ['good'],['bad'],['excellent'],['average'],['good'],['bad'],['excellent'],['bad'],['average'],['good']
]
```

```python
data=pd.DataFrame(data=data,columns=['reviews'])
```

```python
data.head()
```

```python
data.shape
```

```python
categories=[['bad','average','good','excellent']]
```

```python
categories
```

```python
encoder=OrdinalEncoder(categories=categories)
```

```python
encodded_data=encoder.fit_transform(data)
```
