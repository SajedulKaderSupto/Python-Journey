```python
def remove_duplicates(input_list):
    unique_list= []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

```

```python
input_list=[1,2,3,4,2,3,7,8]
remove_duplicates(input_list)
print(f'Real list {input_list}')
print(f'duplicate_list{remove_duplicates(input_list)}')
```

```python
def double_element(input_list):
    double_list=[2*x for x in input_list]
    return double_list
input_list=[1,2,3,4,5,6,7,99,88,1024]
double_element(input_list)
```
