```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
data={'feature1':[1,5,10,4,5],
'feature2':[6,7,8,19,10]
}
df=pd.DataFrame(data)
```

```python
scaller=StandardScaler()
```

```python
standardired_data=scaller.fit_transform(df)
standardized_df=pd.DataFrame(standardired_data,columns=df.columns)
```

```python
print("Original Data")
print(df)
print(standardized_df)
```
